```python
plt.scatter(x, y, color='red')
plt.title('Диаграмма рассеяния')
plt.xlabel('entry x')
plt.ylabel('entry y')
plt.show()
```

```python
plt.plot(x ,y)
plt.scatter(x, y, color='red')
plt.title('График и диаграмма рассеяния')
plt.xlabel('entry x')
plt.ylabel('entry y')
plt.show()
```

#### Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax, причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали. В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам:
##### 'График y1', 'График y2' и т.д.
#### Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
#### Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали. Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, 51)
print(x)
```

```python
y1 = x**2
print(y1)
```

```python
y2 = 2 * x + 0.5
print(y2)
```

```python
y3 = -3 * x - 1.5
print(y3)
```

```python
y4 = np.sin(x)
print(y4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
ax1.set_title('График $y_1$')
ax2.set_title('График $y_2$')
ax3.set_title('График $y_3$')
ax4.set_title('График $y_4$')
ax1.set_xlim([-5, 5])
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
class_list = creditcard['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind='barh')
plt.show()
```

```python
class_list.plot(kind='barh', logx=True)
plt.show()
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.show()
```
